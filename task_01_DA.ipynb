```python
import pandas as pd
import numpy as np
```

```python
raw_data=pd.read_csv("student_performance.csv")
raw_data.head()
```

```python
raw_data.shape
```

```python
raw_data.replace("?", np.nan, inplace = True)
raw_data.head(5)
```

```python
missing_data = raw_data.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")   
```

```python
raw_data.dropna(subset=["StudentID"],inplace=True, axis=0)
```

```python
raw_data.dropna(subset=["Name"],inplace=True, axis=0)
```

```python
raw_data.dropna(subset=["Gender"],inplace=True)
```

```python
avg_attendence=raw_data['AttendanceRate'].astype('float').mean(axis=0)
avg_attendence
raw_data['AttendanceRate'].replace(np.nan,avg_attendence,inplace=True)
```

```python
avg_study_hrs_per_week=raw_data['StudyHoursPerWeek'].astype('float').mean(axis=0)
raw_data['StudyHoursPerWeek'].replace(np.nan,avg_study_hrs_per_week,inplace=True)
```

```python
raw_data['ExtracurricularActivities'].fillna(0,inplace=True)
```

```python
avg_=raw_data['AttendanceRate'].astype('float').mean(axis=0)
avg_attendence
raw_data['AttendanceRate'].replace(np.nan,avg_attendence,inplace=True)
```

```python
# fill numeric columns with median, categorical with mode
for col in raw_data.columns:
    if raw_data[col].dtype in ['int64', 'float64']:
        raw_data[col].fillna(raw_data[col].median(), inplace=True)
    else:
        raw_data[col].fillna(raw_data[col].mode()[0], inplace=True)
```

```python
# Drop duplicate rows
raw_data.drop_duplicates(inplace=True)

```

```python

if 'gender' in raw_data.columns:
    raw_data['gender'] = raw_data['gender'].str.strip().str.lower().replace({
        'm': 'male', 'male': 'male',
        'f': 'female', 'female': 'female'
    })
```

```python
# Clean column names: lowercase, replace spaces with underscores
raw_data.columns = raw_data.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
# Convert numerical columns properly
for col in raw_data.select_dtypes(include=['object']).columns:
    if raw_data[col].str.isnumeric().all():
        raw_data[col] = raw_data[col].astype(int)

# Example: ensure age is int if it exists
if 'age' in raw_data.columns:
    raw_data['age'] = raw_data['age'].astype(int)
```

```python
raw_data.isnull().sum()
```

```python
cleaned_data=raw_data
```

```python
cleaned_data.head()
```

```python
cleaned_data.to_csv("cleaned_student_performance.csv", index=False)
print("Cleaned dataset saved as cleaned_student_performance.csv")
```
